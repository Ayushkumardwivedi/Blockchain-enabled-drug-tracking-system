```python
import google.generativeai as genai
import requests
import IPython.display as display
from PIL import Image

api_key = "AIzaSyBeuZvmNoYIWAUcej3kICpJLa07ikAfJ4E"
genai.configure(api_key=api_key)

model = genai.GenerativeModel(model_name="gemini-1.5-pro",
                              generation_config={
                                  "temperature": 1,
                                  "top_p": 0.95,
                                  "top_k": 64,
                                  "max_output_tokens": 1024,
                                  "response_mime_type":"text/plain",
                              })


def generate_design(style, size, rooms):
    prompt = f"Design a {style}-style home of {size} sq ft with {rooms} rooms. Include layout, color scheme, and furniture suggestions."
    response = model.generate_content(prompt)
    return response.text

def fetch_image(style):
    url = f"https://lexica.art/api/v1/search?q={style} home design"
    response = requests.get(url)
    if response.status_code == 200:
        images = response.json().get('images', [])
        if images:
            return images[0]['src']
    return None

style = input("Enter the design style (e.g., Modern, Rustic): ")
size = input("Enter the size of the home (in sq ft): ")
rooms = input("Enter the number of rooms: ")

design_idea = generate_design(style, size, rooms)
print("\nüí° Home Design Idea:")
print(design_idea)

image_url = fetch_image(style)
if image_url:
    image = Image.open(requests.get(image_url, stream=True).raw)
    display.display(image)
else:
    print("‚ö†Ô∏è No image found for this style.")
```
